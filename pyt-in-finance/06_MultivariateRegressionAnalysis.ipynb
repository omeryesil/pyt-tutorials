# Multivariate Regression Analysis

- Adding more than one explanatory variables to regression testing
- This should provide closer true results than using one explanatory variable
  - Of course the right variables might be used. 
- Multiple beta coefficient and explanatory variable with a constant
  
  
<b>Remember Single Regression</b>
$$
Y_i = \beta_0 + \beta_1X_1 + \varepsilon_1
$$
   
<b>Multivariate Regression</b>
$$
Y_i = \beta_0 + \beta_1X_1 + \beta_2X_2 + \beta_iX_i + \varepsilon_1
$$
- Chec this link for more/detailed info : https://towardsdatascience.com/data-science-simplified-part-5-multivariate-regression-models-7684b0489015 


<b>Remember $R^2$</b>
- Helps us determine how powerfull is the regression
- Values can be between 0-1
- The higher the number of variables, the higher the $R^2$


```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

data = pd.read_excel('samplefiles/housesonsale.xlsx')
```

```python
data
```

## Multivariate Regression

Explanatory Variables to use:
- House Size
- Built Year
- Location Score


```python
x = data[['House Size', 'Built Year', 'Location Score']]
y = data['House Price']
```

```python
x1 = sm.add_constant(x)
reg = sm.OLS(y,x1).fit()
reg.summary()
```

### Independent Varialbes

Variables to use
- House Size
- Built Year



```python
x = data[['House Size', 'Built Year']]
y = data['House Price']
```

```python
x1 = sm.add_constant(x)
reg = sm.OLS(y,x1).fit()

reg.summary()
```
